```python
# loc和iloc的区别
# loc指向的是index，在某些特殊情况下，dataframe的索引并不是唯一的，且不是顺序或逆序的，loc就只是指向特定索引所在数据
# iloc则是指向的位置，与索引是否重复、是否顺序无关，就只代表某一个位置
# 下面举个例子，建立一个series，index存在重复
sample = pd.Series(["cat","dog","duck","rabbit","mice"],index=[0,1,2,3,2])
sample.loc[2]
```

```python
sample.iloc[2]
```

```python
# you see! 和我描述的一样！
```

```python
# [:3]表示从0到3的左闭右开区间
sample.iloc[:3]
```

# 操作数据

```python
Automobile.head().name.str.lower()
# .str.lower()将字符串全都换成小写字母🔡
# 同理，.str.upper()则是全部换成大写字母🔠  
Automobile.head().name.str.upper()
```

```python
# 移除dataframe不需要的行或者列 可以使用drop
subAutomobile = Automobile.head(10)
subAutomobile
```

```python
# 丢掉ford torino🚗所在的信息
# 对于row需要指定index，对于列需要给出column label
subAutomobile.drop(4)
```

```python
subAutomobile
```

```python
# 可见，subAutomobile并没有变化，尽管前一步使用了drop函数，原因在于没有吧drop后的数据重新赋值给subAutomobile
# 解决办法有2种
# 1种是直接赋值
# subAutomobile = subAutomobile.drop(4)
# 另一种是在drop中添加inplace=True
subAutomobile.drop(4,inplace=True)
```

```python
subAutomobile
```

```python
# 哒哒，成功去掉了index=4的row
```

```python
# 下面去掉cylinders这一列
subAutomobile.drop("cylinders",axis=1, inplace=True)
```

```python
subAutomobile
```

```python
# 创造一个新列
seats_column = pd.Series([5,5,5,4,5])
subAutomobile["seats"] = seats_column
subAutomobile
```

```python
subAutomobile.reset_index(drop=True,inplace=True)
# 在这里，使用reset_index来重置数据框的index，因为之前使用drop删除了一行
# reset_index中通过添加drop=True保证旧的index不会成为数据框中新的一列，同时通过inplace=True保证立即修改
```

```python
subAutomobile
# 哒哒，顺利完成✅
```

```python
# 可以看到，表中的数据在seats这一列包含NaN，可以通过手动添加值将其补上
subAutomobile.fillna(5, inplace=True)
subAutomobile
```

```python
# sample，官方文档解释
# Series, DataFrames, and Panels now have a new method: sample(). 
# The method accepts a specific number of rows or columns to return,
# or a fraction of the total number or rows or columns. 
# It also has options for sampling with or without replacement, 
# for passing in a column for weights for non-uniform sampling, 
# and for setting seed values to facilitate replication. 
```

```python
# 理解：对数据进行清洗,frac=1对所有数据进行清洗，frac=0.5表示对一半的数据进行清洗
subAutomobile.sample(frac=1)
```

```python
subAutomobile.sample(frac=0.5)
```

```python
subAutomobile
```

```python
# 可以看到，shuffle并不会将原始数据改变
```

```python
# 现在完成，将数据进行清洗，并且得到清洗后有顺序的index
subAutomobile_shuffled = subAutomobile.sample(frac=1)
subAutomobile_shuffled.reset_index(drop=True, inplace=True)
subAutomobile_shuffled
```

```python
# apply 可以使用函数来操作数据，在这里我们使用lambda匿名函数
subAutomobile_shuffled["displacement"] = subAutomobile_shuffled["displacement"].apply(lambda x: x/2)
subAutomobile_shuffled
```

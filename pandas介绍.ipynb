```python
# sampleÔºåÂÆòÊñπÊñáÊ°£Ëß£Èáä
# Series, DataFrames, and Panels now have a new method: sample(). 
# The method accepts a specific number of rows or columns to return,
# or a fraction of the total number or rows or columns. 
# It also has options for sampling with or without replacement, 
# for passing in a column for weights for non-uniform sampling, 
# and for setting seed values to facilitate replication. 
```

```python
# ÁêÜËß£ÔºöÂØπÊï∞ÊçÆËøõË°åÊ∏ÖÊ¥ó,frac=1ÂØπÊâÄÊúâÊï∞ÊçÆËøõË°åÊ∏ÖÊ¥óÔºåfrac=0.5Ë°®Á§∫ÂØπ‰∏ÄÂçäÁöÑÊï∞ÊçÆËøõË°åÊ∏ÖÊ¥ó
subAutomobile.sample(frac=1)
```

```python
subAutomobile.sample(frac=0.5)
```

```python
subAutomobile
```

```python
# ÂèØ‰ª•ÁúãÂà∞ÔºåshuffleÂπ∂‰∏ç‰ºöÂ∞ÜÂéüÂßãÊï∞ÊçÆÊîπÂèò
```

```python
# Áé∞Âú®ÂÆåÊàêÔºåÂ∞ÜÊï∞ÊçÆËøõË°åÊ∏ÖÊ¥óÔºåÂπ∂‰∏îÂæóÂà∞Ê∏ÖÊ¥óÂêéÊúâÈ°∫Â∫èÁöÑindex
subAutomobile_shuffled = subAutomobile.sample(frac=1)
subAutomobile_shuffled.reset_index(drop=True, inplace=True)
subAutomobile_shuffled
```

```python
# apply ÂèØ‰ª•‰ΩøÁî®ÂáΩÊï∞Êù•Êìç‰ΩúÊï∞ÊçÆÔºåÂú®ËøôÈáåÊàë‰ª¨‰ΩøÁî®lambdaÂåøÂêçÂáΩÊï∞
subAutomobile_shuffled["displacement"] = subAutomobile_shuffled["displacement"].apply(lambda x: x/2)
subAutomobile_shuffled
```

```python
# matplotlib 绘图流程
# 0.导入matplotlib，并设置内联于jupyternotebook
%matplotlib inline
import matplotlib.pyplot as plt

# 1.准备数据
x = [1,2,3,4,5]
y=[11,22,33,44,55]

# 2.设置图形
fig, ax = plt.subplots(figsize=(5,10)) # (width, height)

# 3.绘图
ax.plot(x,y)

# 4.添加样式
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5.保存绘图
fig.savefig("images/sample-plot.png")
```

**思考**：pyplot和matplotlib面向对象方法，哪一种更好？
- 当想要快速的画图时，使用pyplot会好一些
- 当想要以更高级的方式绘图，使用面向对象的方式会好一些

```python
heart_disease.head(10)
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.plot.scatter(x="age",y="chol",c="target")
```

```python
fig, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
```

![anatomy of a matplotlib plot](https://bacon-img.oss-cn-beijing.aliyuncs.com/202409101951343.png)

```python
fig, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
ax.set(xlim=(45,100))
```

```python
fig, ax = plt.subplots(figsize=(10,5))
scatter = ax.scatter(x=over_50["age"], 
           y=over_50["chol"],
           c=over_50["target"])
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

ax.legend(*scatter.legend_elements(), title="Target")
# *表示解包运算符，用于将 scatter.legend_elements() 返回的多个值解包为单独的参数传递给 ax.legend() 函数

ax.axhline(over_50["chol"].mean(), linestyle = "-.", linewidth = 3, color="red")
```

```python
# 绘制组合子图
fig, (ax1,ax2) = plt.subplots(nrows=2,figsize=(10,10),sharex=True)
scatter1 = ax1.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"])
scatter2 = ax2.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"])

# 设置图例
ax1.legend(*scatter1.legend_elements(),title="Target")
ax2.legend(*scatter2.legend_elements(),title="Target")

# 设置样式
ax1.set(title="Heart Disease and Cholesterol Levels",
    #    xlabel="Age",
       ylabel="Cholesterol")
ax2.set(title="Heart Disease and thalach Levels",
       xlabel="Age",
       ylabel="Thalach")

# 设置水平线
ax1.axhline(over_50["chol"].mean(), linestyle = "-.", linewidth = 3, color="red")
ax2.axhline(over_50["thalach"].mean(), linestyle = "--", linewidth = 3, color="blue")

# 设置画布的大标题
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")
```

### `seaborn-v0_8-whitegrid` 感觉是不错的绘图风格！！

cmap = color map，可以在[颜色映射参考](https://matplotlib.org/stable/gallery/color/colormap_reference.html中找到matplotlib的颜色映射参考)中选择你喜欢的预设样式

```python
# 设置绘图样式
plt.style.use("seaborn-v0_8-whitegrid")

# 准备
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
           y=over_50["chol"],
           c=over_50["target"],
           cmap="viridis")

# 设置文字
ax.set(title="Heart disease plot",
       xlabel="Age",
       ylabel="Chol")

# 设置图例
ax.legend(*scatter.legend_elements(), title="Target")

# 设置水平基线
ax.axhline(y=over_50["chol"].mean(),linestyle="--")
# 出图
```

```python
# 绘制组合子图
fig, (ax1,ax2) = plt.subplots(nrows=2,figsize=(10,10),sharex=True)
scatter1 = ax1.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="plasma")
scatter2 = ax2.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="magma")

# 设置图例
ax1.legend(*scatter1.legend_elements(),title="Target")
ax2.legend(*scatter2.legend_elements(),title="Target")

# 设置样式
ax1.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol",
       xlim=[50,80])
ax2.set(title="Heart Disease and thalach Levels",
       xlabel="Age",
       ylabel="Thalach",
       ylim=[60,200])

# 设置水平线
ax1.axhline(over_50["chol"].mean(), linestyle = "-.", linewidth = 3, color="red")
ax2.axhline(over_50["thalach"].mean(), linestyle = "--", linewidth = 3, color="blue")

# 设置画布的大标题
fig.suptitle("Heart Disease Analysis", fontsize=20, fontweight="bold")

# 保存绘图
fig.savefig("./images/heart-disease-2-param.png")
```
